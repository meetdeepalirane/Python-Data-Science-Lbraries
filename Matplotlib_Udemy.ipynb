```python

# Functional Method


import numpy as np
x=np.linspace(0,5,11)
y=x**2
plt.xlabel('X label')
plt.ylabel('Y Label')
plt.title('Title')
plt.plot(x,y,'r-')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g-')


plt.subplot(1,2,2)
plt.plot(y,x,'b-')
```

```python
# Object Oriented Method

fig=plt.figure()
axes=fig.add_axes([0.6,0.5,0.6,0.6])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)

#to fix the overlaps
plt.tight_layout()


```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.plot(x,y)
#axes is array of list of  matplotlib objects
axes

#you can iterate through these axes
for current_axes in axes:
    current_axes.plot(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)
fig,axes=plt.subplots(figsize=(8,7))
axes.plot(X,C,marker="*",markersize=4,markerfacecolor="yellow",markeredgecolor="green",label="COSINE")
axes.plot(X,S,'*-b',markersize=4,label="SIN")

axes.set_xlabel("Number")
axes.set_ylabel("Value")


#axes.set_xlim(1,2)
axes.legend(loc=0)

#axes.spines['right'].set_color('none')




plt.tight_layout()
```

```python
# Regular Plot


import numpy as np


X=np.linspace(-np.pi,np.pi,256,endpoint=True)
Y=np.sin(2*X)

fig,ax=plt.subplots()
ax.plot(X,Y+1,alpha=0.8)
ax.plot(X,Y-1,alpha=0.8)
```

```python

#Scatter Plot


#The scatter() function plots one dot for each observation. It needs two arrays of the same length, 
#one for the values of the x-axis, and one for values on the y-axis:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x,y,c=colors,cmap='viridis')      
 #cmap---color map by default cmap='viridis'
plt.colorbar()
```

```python
#bar
x=np.array(['A','B',"C",'D','E'])
y = np.array([3, 8, 1, 10,15])
plt.bar(x,y,color='green',width=0.5)
```

```python
plt.barh(x,y)            #horizontal bar
```

```python
#Histogram
x=np.random.normal(10,100,250)
plt.hist(x)
```

```python
# Pie chart
x=np.array([35, 30, 35, 15])
la=['India','USA','UK','France']#giving labels
plt.pie(x,labels=la,startangle=90,shadow=True)             #startangle sets starting angle bydefaults its x axis and degree 0


#explode()--one of the wedges to standout..... explode(0.2,0,0,0)...only first will explode rest ll stand still
plt.show()
```

```python
x=np.array([35, 30, 35, 15])
la=['India','USA','UK','France']
plt.pie(x,labels=la,startangle=90,explode=[0.2,0,0,0],shadow=True)             
#explode()--one of the wedges to standout..... explode(0.2,0,0,0)...only first will explode rest ll stand still
plt.legend(title="Countries")
plt.show()
```
